# Demo 2：AI 翻譯機（英文 → 中文）

> Jupyter Notebook 版本 — 第二個 AI，學會「指定模型」

## 這個 notebook 要做什麼？

你給 AI 一句英文，AI 翻成中文還給你。

這個任務叫做 **Machine Translation（機器翻譯）**。Google 翻譯、DeepL 都是這類技術，只是規模大很多。

## 跟 Demo 1 差在哪？

Demo 1 我們讓 HuggingFace 自己選模型。

這次我們要**自己指定模型**。為什麼？因為翻譯有**方向**（英→中、中→英、英→日…），AI 不能猜你想翻哪個方向。

## 第一次跑會發生什麼？

- 「建立翻譯機器」那格會下載一個英翻中模型，約 300 MB，要等 1～2 分鐘

## 步驟 1：借工具（跟 Demo 1 一樣）

```python
from transformers import pipeline
```

## 步驟 2：建立翻譯機器（指定模型）

這次的 `pipeline()` 多了一個 `model=...` 參數：

- `"translation"` —— 任務名稱
- `model="Helsinki-NLP/opus-mt-en-zh"` —— **指定**用哪個模型

**模型名字怎麼拆？**

| 部分 | 意思 |
|---|---|
| `Helsinki-NLP` | 發佈者（赫爾辛基大學 NLP 團隊） |
| `/` | 分隔符 |
| `opus-mt-en-zh` | 模型名（OPUS 資料集訓練、英→中） |

格式很像 GitHub 的 `用戶名/專案名`。你可以去 https://huggingface.co/Helsinki-NLP 看他們發的所有翻譯模型（有超過 1000 個，涵蓋幾百種語言對）。

> ⏳ 第一次跑會下載模型（約 300 MB），耐心等 `[*]` 變數字。

```python
translator = pipeline(
    "translation",
    model="Helsinki-NLP/opus-mt-en-zh",
)
```

## 步驟 3：準備幾句英文

```python
english_sentences = [
    "Hello, world!",
    "I am learning artificial intelligence.",
    "Taiwan is a beautiful island.",
    "Programming is easier than people think.",
]
```

## 步驟 4：一次翻譯整批

把整個清單丟給 AI，會比一句一句翻快很多（GPU/CPU 同時處理整批比較有效率）。

```python
results = translator(english_sentences)
results
```

AI 回傳的每個結果都有一個 `translation_text` 欄位，裡面就是譯文。

## 步驟 5：印出對照表

```python
print("\n=== AI 英翻中結果 ===\n")
for english, result in zip(english_sentences, results):
    chinese = result["translation_text"]
    print(f"原文：{english}")
    print(f"譯文：{chinese}\n")
```

## ⚠️ 注意 AI 翻錯的地方

你應該會看到最後一句 `Programming is easier than people think.` 被翻成「**規劃**比人們想象的容易」，而不是「**程式設計**比人們想象的容易」。

**這是真實的 AI 缺點**：開源小模型在「一字多義」時常常猜錯。商用大模型（ChatGPT、DeepL）通常會翻對。

這也是為什麼學 AI 要學「**怎麼選模型**」，不只是「**怎麼跑模型**」。

## 動手玩玩看

### 玩法 1：換句子

在下面的 `my_english` 加你自己的英文：

```python
my_english = [
    "My cat is sleeping on the keyboard.",
    "This homework is too hard.",
    # ← 加你自己的句子
]

for en, result in zip(my_english, translator(my_english)):
    print(f"{en}\n  → {result['translation_text']}\n")
```

### 玩法 2：換方向（中翻英）

把模型換成 `opus-mt-zh-en`，餵中文進去：

```python
translator_zh_en = pipeline(
    "translation",
    model="Helsinki-NLP/opus-mt-zh-en",
)

chinese_sentences = [
    "今天天氣真好。",
    "我喜歡吃珍珠奶茶。",
    "人工智慧正在改變世界。",
]

for zh, result in zip(chinese_sentences, translator_zh_en(chinese_sentences)):
    print(f"{zh}\n  → {result['translation_text']}\n")
```

### 玩法 3：換語言（你自己挑）

去 HuggingFace 搜尋 `opus-mt-en-XX`（`XX` 是兩字母語言代號），常見的有：

- `opus-mt-en-ja` — 英翻日
- `opus-mt-en-fr` — 英翻法
- `opus-mt-en-de` — 英翻德
- `opus-mt-en-ko` — 英翻韓

把上面 `translator = pipeline(...)` 的 `model=` 換掉，再餵英文進去試試。
