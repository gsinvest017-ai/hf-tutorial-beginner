# Demo 1：你的第一個 AI（情感分析）

> Jupyter Notebook 版本 — 一格一格慢慢跑，看 AI 怎麼運作

## 這個 notebook 要做什麼？

你給 AI 一句話，AI 告訴你這句話是「**正面（開心）**」還是「**負面（難過）**」。

這個任務叫做 **Sentiment Analysis（情感分析）**，是 AI 領域最入門的任務之一。

## 怎麼用這個 notebook？

1. 從上往下，**一格一格按 `Shift + Enter`** 執行
2. 每跑完一格，就看一下它的輸出，再跑下一格
3. 跑到最後一格，你會看到 AI 的判斷結果
4. 後面有「**動手玩玩看**」區，你可以改裡面的句子再跑

**第一次跑會發生什麼？**
- 第 2 格（建立分析機器）會「下載」一個 AI 模型，約 250 MB，要等 1～2 分鐘
- 之後再跑就不用下載了，會很快

## 步驟 1：借工具

從 `transformers` 這個套件，借一個叫 `pipeline` 的工具來用。

`pipeline` 是 HuggingFace 最簡單的入口，就像「一鍵 AI」按鈕。

````python
from transformers import pipeline
````

## 步驟 2：建立「情感分析機器」

`"sentiment-analysis"` 是告訴 HuggingFace：「我要的是**情感分析**這項功能」。

HuggingFace 會**自動幫你下載**一個適合的 AI 模型。

> ⏳ **第一次跑這格會比較久**（約 1～2 分鐘），因為要下載模型。請耐心等左邊的 `[*]` 變成數字。

````python
classifier = pipeline("sentiment-analysis")
````

## 步驟 3：準備幾句要分析的話

````python
sentences = [
    "I love this movie, it's amazing!",      # 「我超愛這部電影！」→ 應該是正面
    "This is the worst day of my life.",     # 「今天是我人生最爛的一天」→ 應該是負面
    "The food was okay, nothing special.",   # 「東西還好，沒什麼特別」→ 中性偏負面
]
````

## 步驟 4：把句子交給 AI

AI 會回傳一個結果清單。每個結果都有 `label`（標籤）跟 `score`（信心度）。

````python
results = classifier(sentences)
results
````

看出來了嗎？AI 給每句話一個 `label`（POSITIVE 或 NEGATIVE）跟 `score`（0～1 之間的信心度）。

## 步驟 5：把結果印得好看一點

用一個小迴圈把原句跟 AI 的判斷對照印出來。

````python
print("\n=== AI 情感分析結果 ===\n")
for sentence, result in zip(sentences, results):
    label = result["label"]              # POSITIVE 或 NEGATIVE
    score = result["score"]              # AI 的信心度，0～1 之間
    emoji = "😊" if label == "POSITIVE" else "😢"
    print(f"{emoji} 句子：{sentence}")
    print(f"   AI 判斷：{label}（信心 {score:.1%}）\n")
````

🎉 **恭喜！你剛剛跑了你人生第一個 AI**。

## 動手玩玩看

把下面的 `my_sentences` 改成**你自己**想分析的句子，再跑這格看看：

- 你今天的心情
- 一句你最近說過的話
- 一句你想知道 AI 怎麼看的話

**注意**：這個模型只認**英文**。中文版要等下個 notebook（`02_translate.ipynb`）才會用到。

````python
my_sentences = [
    "I just finished my homework, finally!",
    "Why does Monday come so fast...",
    # ← 在這裡加你自己的句子
]

for sentence, result in zip(my_sentences, classifier(my_sentences)):
    emoji = "😊" if result["label"] == "POSITIVE" else "😢"
    print(f"{emoji} {sentence} → {result['label']} ({result['score']:.1%})")
````

## 你剛剛做了什麼？

**3 個關鍵步驟**，就跑出了一個真的 AI：

```python
from transformers import pipeline           # ① 借工具
classifier = pipeline("sentiment-analysis") # ② 開機器
results = classifier(["I love this!"])      # ③ 用機器
```

背後 HuggingFace 幫你做了：

1. 下載一個叫 **DistilBERT** 的神經網路（幾億個參數）
2. 把你的句子切成「詞」（tokenization）
3. 把詞變成數字（embedding）
4. 餵進神經網路跑一輪（forward pass）
5. 把輸出的數字翻譯回 `POSITIVE / NEGATIVE`

這 5 步如果你自己寫會寫到吐血。**`pipeline()` 幫你一行搞定** —— 這就是「**抽象層**」的威力。
